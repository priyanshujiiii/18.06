# Matrix Spaces; Rank-1; Small World Graphs











## 1) Derivation

1. **Matrix Spaces**:

   * A set of all $m \times n$ matrices forms a vector space under matrix addition and scalar multiplication.
   * Subspaces include: row space, column space, null space.

2. **Rank-1 Matrices**:

   * A matrix $A$ has rank 1 if it can be written as an outer product of two vectors: $$ A = u v^T, \quad u \in \mathbb{R}^m, \; v \in \mathbb{R}^n $$


   * Rank 1 means all rows are multiples of one row, or all columns are multiples of one column.

3. **Small World Graphs**:

   * In graph theory, a **small-world network** is one where most nodes are not neighbors but can be reached by a small number of steps.
   * Constructed often using the **Wattsâ€“Strogatz model**.
   * Adjacency matrices of such graphs often have interesting rank and eigenvalue structures.

## 2) Definitions

* **Matrix Space**: Vector space formed by all $m \times n$ matrices.
* **Rank-1 Matrix**: Matrix whose column/row space is 1-dimensional.
* **Small World Graph**: A graph with high clustering coefficient and short average path length.


## 3) Computation (Step-by-Step)


```python
import numpy as np
from sympy import Matrix

# Example: Constructing a rank-1 matrix
u = Matrix([1, 2, 3])
v = Matrix([4, 5, 6])
A = u * v.T

print("Rank-1 Matrix A = u v^T:")
display(A)

# Step 1: Rank of A
print("\nStep 1: Rank of A =", A.rank())

# Step 2: Column space of A
col_space = A.columnspace()
print("\nStep 2: Column Space Basis:")
for c in col_space:
    print(c)

# Step 3: Row space of A
row_space = A.rowspace()
print("\nStep 3: Row Space Basis:")
for r in row_space:
    print(r)

# Step 4: Null space of A
null_space = A.nullspace()
print("\nStep 4: Null Space Basis:")
for n in null_space:
    print(n)

```


## 4) Plotting
### (a) Rank-1 Matrix Visualization

```python
import matplotlib.pyplot as plt

# Apply transformation A on random points
points = np.random.randn(2, 100)
A2 = np.array([[2],[1]]) @ np.array([[3,4]])   # Rank-1 matrix in 2D
transformed = A2 @ points

plt.figure(figsize=(6,6))
plt.scatter(points[0], points[1], alpha=0.3, label="Original Points")
plt.scatter(transformed[0], transformed[1], alpha=0.7, label="After Rank-1 Transform")
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.legend()
plt.title("Effect of Rank-1 Transformation")
plt.show()
```

### (b) Small-World Graph Visualization

```python
import networkx as nx

# Create small world graph using Wattsâ€“Strogatz model
G = nx.watts_strogatz_graph(20, 4, 0.3)  # 20 nodes, each connected to 4, rewiring prob=0.3

plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500)
plt.title("Small World Graph (Watts-Strogatz Model)")
plt.show()

# Adjacency matrix rank
A_graph = Matrix(nx.adjacency_matrix(G).todense())
print("\nRank of Small World Graph Adjacency Matrix:", A_graph.rank())
```

## 5) Use Cases

* **Matrix Spaces**:

  * Foundation for linear algebra, data science, and ML.
  * Helps in understanding transformations and subspace structures.

* **Rank-1 Matrices**:

  * Used in **matrix factorization**, **recommendation systems** (low-rank approximations).
  * Image compression (approximate image by low-rank matrices).
  * PCA (principal components yield rank-1 approximations).

* **Small World Graphs**:

  * Model **social networks**, **neural networks**, **internet connectivity**.
  * Explain why real-world graphs have **short paths** despite sparse connections.
  * Used in epidemiology (disease spread modeling).

**Key Values They Produce**:

* Rank â†’ measures degrees of freedom / independent information.
* Basis of row/column/null spaces â†’ minimal representation of matrix behavior.
* Graph properties â†’ connectivity, efficiency of communication, robustness.
