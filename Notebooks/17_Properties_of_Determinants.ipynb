# Properties of Determinants

## 1) Definition

The **determinant** of a square matrix $A \in \mathbb{R}^{n \times n}$, denoted $\det(A)$ or $|A|$, is a scalar value that encodes important properties of the matrix:

- $\det(A) = 0$ if and only if $A$ is singular (non-invertible).
- $\det(A)$ represents the **volume scaling factor** of the linear transformation defined by $A$.

For a $2 \times 2$ matrix:

$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}, \quad \det(A) = ad - bc
$$

For $3 \times 3$:

$$
A = \begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix}, \quad
\det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$


## 2) Derivation

**Properties of Determinants:**

1. **Linearity:** The determinant is linear in each row (or column) separately.
2. **Swapping Rows:** Swapping two rows multiplies the determinant by $-1$.
3. **Scaling Rows:** Multiplying a row by scalar $k$ multiplies $\det(A)$ by $k$.
4. **Row Addition:** Adding a multiple of one row to another does not change the determinant.
5. **Triangular Matrix:** The determinant of a triangular matrix (upper or lower) is the product of its diagonal entries:
$$
\det\begin{bmatrix} a_{11} & * & * \\ 0 & a_{22} & * \\ 0 & 0 & a_{33} \end{bmatrix} = a_{11} a_{22} a_{33}
$$
6. **Product of Matrices:**
$$
\det(AB) = \det(A) \cdot \det(B)
$$
7. **Inverse Matrix:** If $A$ is invertible:
$$
\det(A^{-1}) = \frac{1}{\det(A)}
$$
8. **Transpose:**
$$
\det(A^T) = \det(A)
$$


## 3) Computation Steps

**Step 1:** Identify the square matrix $A$.

**Step 2:** For a $2\times2$ matrix:
$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}, \quad \det(A) = ad - bc
$$

**Step 3:** For a $3\times3$ matrix:
$$
\det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$

**Step 4:** For larger matrices, use JAX's `jnp.linalg.det(A)` to compute efficiently.

**Step 5:** Apply properties:
- Swapping rows → multiply by -1
- Scaling rows → multiply determinant by scalar
- Triangular matrices → product of diagonal


```python
import jax.numpy as jnp

# Example matrices
A2 = jnp.array([[2., 3.],
                [1., 4.]])
A3 = jnp.array([[1., 2., 3.],
                [0., 4., 5.],
                [1., 0., 6.]])

# Compute determinants
det_A2 = jnp.linalg.det(A2)
det_A3 = jnp.linalg.det(A3)

print("Matrix A2:\n", A2)
print("Determinant of A2:", det_A2)

print("\nMatrix A3:\n", A3)
print("Determinant of A3:", det_A3)
```

## 4) Plotting

```python
import matplotlib.pyplot as plt

# 2D vectors as columns of matrix
v1 = jnp.array([2., 1.])
v2 = jnp.array([3., 4.])

# Original parallelogram
origin = jnp.zeros(2)
plt.figure(figsize=(6,6))
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# Fill parallelogram
plt.fill([0, v1[0], v1[0]+v2[0], v2[0]],
         [0, v1[1], v1[1]+v2[1], v2[1]], color='y', alpha=0.3)

# Compute determinant (area)
A2 = jnp.column_stack([v1, v2])
area = jnp.abs(jnp.linalg.det(A2))

plt.xlim(0,5)
plt.ylim(0,5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title(f"Parallelogram spanned by v1 and v2\nArea = |det(A)| = {area:.2f}")
plt.show()
```

## 5) Use Cases of Determinants

1. **Invertibility:** $\det(A) \neq 0 \Rightarrow A$ is invertible.
2. **Volume/Area Scaling:** Determinant gives the scaling factor of volume under the linear transformation defined by $A$.
3. **Eigenvalues:** Used in characteristic polynomial: $\det(A - \lambda I) = 0$.
4. **Change of Variables in Integrals:** Jacobian determinant in multivariable calculus.
5. **Solving Systems:** Determinants appear in Cramer's rule for solving linear equations.
