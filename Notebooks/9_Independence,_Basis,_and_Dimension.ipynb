# Independence, Basis, and Dimension

# 1) Derivation

* A set of vectors ${v_1, v_2, \dots, v_k}$ in $\mathbb{R}^n$ is linearly independent if:
$$ c_1 v_1 + c_2 v_2 + \cdots + c_k v_k = 0 \quad \implies \quad c_1 = c_2 = \cdots = c_k = 0 $$

* Otherwise, they are linearly dependent.

* A basis of a vector space $V$ is a linearly independent set of vectors that span $V$.

* The dimension of $V$ is the number of vectors in a basis:
$$ \dim(V) = \text{number of basis vectors} $$

# 2) Definitions
* Linear Independence: No vector can be written as a linear combination of others.
* Basis: Minimal spanning set of vectors.
* Dimension: Size of the basis (degrees of freedom of the space).

#  3) Computation (Step-by-Step)

```python
import numpy as np
from sympy import Matrix

# Example vectors in R^3
v1 = Matrix([1, 2, 3])
v2 = Matrix([2, 4, 6])
v3 = Matrix([1, 0, 1])

# Step 1: Form matrix with these as columns
A = Matrix.hstack(v1, v2, v3)
print("Matrix with column vectors:")
display(A)

# Step 2: Row reduce to check independence
rref_matrix, pivots = A.rref()
print("\nStep 2: Row Reduced Echelon Form (RREF):")
display(rref_matrix)
print("Pivot columns:", pivots)

# Step 3: Basis of the span
basis = A.columnspace()
print("\nStep 3: Basis for span(v1, v2, v3):")
for b in basis:
    print(b)

# Step 4: Dimension
dimension = len(basis)
print("\nStep 4: Dimension of span(v1, v2, v3):", dimension)
```

# 4) Plotting
We can visualize vectors in $\mathbb{R}^3$, showing dependency vs independence.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Original vectors
vecs = [np.array(v1, dtype=float), np.array(v2, dtype=float), np.array(v3, dtype=float)]
colors = ['r','g','b']
labels = ['v1','v2','v3']

for v,c,l in zip(vecs,colors,labels):
    ax.quiver(0,0,0, v[0], v[1], v[2], color=c, label=l)

ax.set_xlim([-2,4])
ax.set_ylim([-2,5])
ax.set_zlim([-1,7])
ax.set_title("Vectors in R^3")
ax.legend()
plt.show()
```

* Here $v_1$ and $v_2$ are dependent ($v_2 = 2 v_1$).
* The basis is ${v_1, v_3}$, so $\dim(\text{span}{v_1,v_2,v_3}) = 2$.

# 5) Use Cases
* Linear Independence: Detect redundant information (e.g., redundant features in ML).
* Basis: Provides the minimal representation of a space.
* Dimension: Measures degrees of freedom / complexity of a system.

Applications:

* Data compression (remove dependent vectors).
* Principal Component Analysis (PCA)  finding basis of data subspace.
* Computer graphics (basis vectors define coordinate frames).
* Control theory and physics (degrees of freedom).

Key Values Produced:

* Whether vectors are independent or dependent.
* A basis set (simplified representation).
* The dimension of the space they span.
