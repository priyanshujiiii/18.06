# 1) Derivation
Given a matrix $A \in \mathbb{R}^{m \times n}$:

Column Space (Range):
The set of all linear combinations of the columns of $A$.
$$ \text{Col}(A) = { A x : x \in \mathbb{R}^n } $$

Null Space (Kernel):
The set of all solutions to the homogeneous system $Ax = 0$.
$$ \text{Null}(A) = { x \in \mathbb{R}^n : A x = 0 } $$

Fundamental Theorem of Linear Algebra (Dimension Relationship):
$$ \text{rank}(A) + \text{nullity}(A) = n $$
where $n$ = number of columns of $A$.

# 2) Definitions
Column Space: Subspace of $\mathbb{R}^m$ spanned by the columns of $A$.
Null Space: Subspace of $\mathbb{R}^n$ consisting of all $x$ such that $Ax=0$.

# 3) Computation

```python
import numpy as np
from sympy import Matrix

# Example matrix
A = np.array([[2, 4, 1],
              [0, 0, 1],
              [0, 0, 0]], dtype=float)

# Convert to sympy Matrix for exact computations
M = Matrix(A)

# Column space basis
col_space = M.columnspace()
print("Column Space Basis:")
for v in col_space:
    print(v)

# Null space basis
null_space = M.nullspace()
print("\nNull Space Basis:")
for v in null_space:
    print(v)

# Rank and nullity
print("\nRank:", M.rank())
print("Nullity:", A.shape[1] - M.rank())
```

#4) Plotting

We can visualize the column space (subspace in $\mathbb{R}^2$ or $\mathbb{R}^3$) and the null space solutions.

```python


import matplotlib.pyplot as plt
# Column vectors
col_vecs = np.array(M.columnspace(), dtype=float)

fig = plt.figure(figsize=(10,5))

# --- Plot Column Space ---
ax1 = fig.add_subplot(121, projection='3d')
ax1.quiver(0,0,0, col_vecs[0][0], col_vecs[0][1], col_vecs[0][2], color='r', label="Col1")
if len(col_vecs) > 1:
    ax1.quiver(0,0,0, col_vecs[1][0], col_vecs[1][1], col_vecs[1][2], color='b', label="Col2")
ax1.set_title("Column Space")
ax1.legend()

# --- Plot Null Space ---
ax2 = fig.add_subplot(122, projection='3d')
null_vecs = np.array(M.nullspace(), dtype=float)
if len(null_vecs) > 0:
    ax2.quiver(0,0,0, null_vecs[0][0], null_vecs[0][1], null_vecs[0][2], color='g', label="NullVec")
ax2.set_title("Null Space")
ax2.legend()

plt.show()
```

#5) Use Cases

Column Space:

Represents the range of possible outputs $Ax$.


*   In data science: tells us what transformations preserve information.
*   In machine learning: relates to feature span and linear dependence.

Null Space:

Represents redundant directions in input that map to zero.
* In linear regression: corresponds to non-unique solutions.
* In control theory: useful for analyzing stability and constraints.

Key Value They Produce:

Rank (dimension of column space) → effective degrees of freedom.


Nullity (dimension of null space) → redundancy in representation.
