# Projections onto Subspaces

## 1) Definition

Let $\mathbf{v} \in \mathbb{R}^n$ be a vector and $U$ be a subspace of $\mathbb{R}^n$ spanned by an orthonormal basis $\{\mathbf{u}_1, \mathbf{u}_2, \dots, \mathbf{u}_k\}$.  

The **orthogonal projection** of $\mathbf{v}$ onto $U$ is:

$$
\text{proj}_U(\mathbf{v}) = (\mathbf{v} \cdot \mathbf{u}_1)\mathbf{u}_1 + (\mathbf{v} \cdot \mathbf{u}_2)\mathbf{u}_2 + \dots + (\mathbf{v} \cdot \mathbf{u}_k)\mathbf{u}_k
$$

The projection lies in the subspace $U$, and the residual vector $\mathbf{v} - \text{proj}_U(\mathbf{v})$ is orthogonal to $U$.


## 2) Derivation

1. Let $U$ be spanned by orthonormal basis $\{\mathbf{u}_1, \dots, \mathbf{u}_k\}$.
2. We want to find $\mathbf{p} \in U$ such that $\mathbf{v} - \mathbf{p}$ is orthogonal to $U$.
3. Orthogonality condition:
$$
(\mathbf{v} - \mathbf{p}) \cdot \mathbf{u}_i = 0, \quad \forall i = 1, \dots, k
$$
4. Let $\mathbf{p} = \sum_{i=1}^{k} c_i \mathbf{u}_i$. Then:
$$
(\mathbf{v} - \sum_{i=1}^{k} c_i \mathbf{u}_i) \cdot \mathbf{u}_j = 0 \quad \Rightarrow \quad c_j = \mathbf{v} \cdot \mathbf{u}_j
$$
5. So the projection formula becomes:
$$
\text{proj}_U(\mathbf{v}) = \sum_{i=1}^{k} (\mathbf{v} \cdot \mathbf{u}_i) \mathbf{u}_i
$$


## 3) Computation Steps

**Step 1:** Identify the vector $\mathbf{v}$ and subspace $U$ with orthonormal basis $\{\mathbf{u}_1, \dots, \mathbf{u}_k\}$.

**Step 2:** Compute the dot products:
$$
c_i = \mathbf{v} \cdot \mathbf{u}_i
$$

**Step 3:** Multiply each basis vector by its coefficient and sum:
$$
\text{proj}_U(\mathbf{v}) = \sum_{i=1}^{k} c_i \mathbf{u}_i
$$

**Step 4:** Compute the residual vector orthogonal to $U$:
$$
\mathbf{r} = \mathbf{v} - \text{proj}_U(\mathbf{v})
$$


```python
# Orthogonal projection of a vector onto a subspace
import numpy as np

# Orthonormal basis for subspace U
u1 = np.array([1, 0])
u2 = np.array([0, 1])  # 2D standard basis

# Vector to project
v = np.array([2, 3])

# Compute projection
proj_v_on_U = np.dot(v, u1)*u1 + np.dot(v, u2)*u2

# Compute residual (orthogonal component)
residual = v - proj_v_on_U

print("Vector v:", v)
print("Projection of v onto U:", proj_v_on_U)
print("Residual vector (v - proj_U(v)):", residual)
```

#4) Plotting

```python
# Plot the vector, projection, and residual
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
origin = np.array([0,0])

# Original vector
plt.quiver(*origin, *v, color='r', angles='xy', scale_units='xy', scale=1, label='v')

# Projection
plt.quiver(*origin, *proj_v_on_U, color='b', angles='xy', scale_units='xy', scale=1, label='proj_U(v)')

# Residual (starts at tip of projection)
plt.quiver(*proj_v_on_U, *residual, color='g', angles='xy', scale_units='xy', scale=1, label='v - proj_U(v)')

plt.xlim(0, 4)
plt.ylim(0, 4)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Vector v, Projection onto Subspace U, and Residual")
plt.show()
```

## 5) Use Cases of Projections onto Subspaces

1. **Least Squares Regression:** Projection of observed data onto the column space of features minimizes the squared error.
2. **Dimensionality Reduction:** PCA projects high-dimensional data onto a lower-dimensional principal subspace.
3. **Signal Processing:** Extracting components of a signal along orthogonal directions.
4. **Computer Graphics:** Projecting points onto planes or surfaces for rendering.
5. **Optimization:** Resolving vectors along constrained directions in engineering or physics.
