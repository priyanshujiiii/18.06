# The Four Fundamental Subspaces


## 1) Derivation

For a matrix $A \in \mathbb{R}^{m \times n}$, the **Four Fundamental Subspaces** arise from solving $Ax = b$ and $A^Ty = 0$:

1. **Column Space** $\text{Col}(A)$:
   $$ \text{Col}(A) = { Ax : x \in \mathbb{R}^n } \subseteq \mathbb{R}^m $$
   Represents all possible outputs of $Ax$.

2. **Null Space** $\text{Null}(A)$:
   $$ \text{Null}(A) = { x \in \mathbb{R}^n : Ax = 0 } $$
   Represents inputs that map to zero.

3. **Row Space** $\text{Row}(A)$:
   $$ \text{Row}(A) = \text{Col}(A^T) \subseteq \mathbb{R}^n $$
   Span of rows of $A$.

4. **Left Null Space** $\text{Null}(A^T)$:
   $$ \text{Null}(A^T) = { y \in \mathbb{R}^m : A^Ty = 0 } $$
   Orthogonal complement of column space.

**Fundamental Theorem of Linear Algebra**:

* $\dim(\text{Col}(A)) = \dim(\text{Row}(A)) = \text{rank}(A)$
* $\dim(\text{Null}(A)) + \dim(\text{Row}(A)) = n$
* $\dim(\text{Col}(A)) + \dim(\text{Null}(A^T)) = m$





## 2) Definitions

* **Column Space**: All linear combinations of columns of $A$.
* **Row Space**: All linear combinations of rows of $A$.
* **Null Space**: All solutions to $Ax = 0$.
* **Left Null Space**: All solutions to $A^Ty = 0$.



## 3) Computation (Step-by-Step)

```python
import numpy as np
from sympy import Matrix

# Example matrix
A = Matrix([[1, 2, 3],
            [4, 5, 6],
            [7, 8, 9]])

print("Matrix A:")
display(A)

# Step 1: Column space
col_space = A.columnspace()
print("\nStep 1: Column Space Basis:")
for v in col_space:
    print(v)

# Step 2: Row space
row_space = A.rowspace()
print("\nStep 2: Row Space Basis:")
for v in row_space:
    print(v)

# Step 3: Null space
null_space = A.nullspace()
print("\nStep 3: Null Space Basis:")
for v in null_space:
    print(v)

# Step 4: Left null space
left_null_space = A.T.nullspace()
print("\nStep 4: Left Null Space Basis:")
for v in left_null_space:
    print(v)

# Step 5: Rank and dimensions
print("\nRank(A):", A.rank())
print("Dimension of Col(A):", len(col_space))
print("Dimension of Row(A):", len(row_space))
print("Dimension of Null(A):", len(null_space))
print("Dimension of Null(A^T):", len(left_null_space))
```

## 4) Plotting

We can visualize the column space and null space for a $3 \times 3$ example.



```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,5))

# --- Column Space ---
ax1 = fig.add_subplot(121, projection='3d')
for v in col_space:
    vec = np.array(v, dtype=float)
    ax1.quiver(0,0,0, vec[0], vec[1], vec[2], color='r')
ax1.set_title("Column Space (Span of Columns)")

# --- Null Space ---
ax2 = fig.add_subplot(122, projection='3d')
for v in null_space:
    vec = np.array(v, dtype=float)
    ax2.quiver(0,0,0, vec[0], vec[1], vec[2], color='b')
ax2.set_title("Null Space (Solutions to Ax=0)")

plt.show()
```


* The **column space** shows directions that $Ax$ can output.
* The **null space** shows input directions that collapse to zero.

(Row space and left null space can also be visualized in $\mathbb{R}^2$ or $\mathbb{R}^3$, but harder for higher dimensions.)

## 5) Use Cases

* **Column Space**: Determines feasible outputs; used in regression and transformations.
* **Row Space**: Captures constraints among variables.
* **Null Space**: Captures redundancy, important in optimization and least-squares problems.
* **Left Null Space**: Detects inconsistencies (residuals in $Ax=b$).

**Applications**:

* Data science: Rank, dimension reduction, PCA.
* Engineering: Solving under/over-determined systems.
* Control systems: Controllability and observability.
* Computer graphics: Transformations and projections.

**Key Values Produced**:

* Rank (dimension of row/column space).
* Nullity (dimension of null/left null space).
* Basis vectors for each subspace.
